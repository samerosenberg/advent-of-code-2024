## Part 1

```python
def load_grid(file_path):
    with open(file_path, 'r') as f:
        return [list(line.strip()) for line in f]

def find_guard(grid):
    for i in range(len(grid)):
        for j in range(len(grid[i])):
            if grid[i][j] == "^":
                return i, j

def move_guard(grid):
    row, col = find_guard(grid)
    direction = 3

    directions = [
        (0, 1),   # Right
        (1, 0),   # Down
        (0, -1),  # Left
        (-1, 0),  # Up
    ]
    
    left_area = False
    area_traveled = 1

    while not left_area:
        new_row, new_col = row + directions[direction][0], col + directions[direction][1]
        if 0 <= new_row < len(grid) and 0 <= new_col < len(grid[0]):
            if grid[new_row][new_col] == "#":
                direction = (direction + 1) % 4
                continue
            else:
                if grid[new_row][new_col] != "X":
                    area_traveled += 1
                grid[row][col] = "X"
                row, col = new_row, new_col
        else:
            left_area = True
    
    return grid, area_traveled

grid = load_grid("input.txt")

final_grid, area_traveled = move_guard(grid)

# print(area_traveled)
# for row in final_grid:
#     print("".join(row))


```
