# Day 1: Historian Hysteria

[Day 1](https://adventofcode.com/2024/day/1)

## Part 1

```python
historian_one = []
historian_two = []

with open('input.txt', 'r') as file:
    for line in file:
        location_ids = line.split();
        historian_one.append(int(location_ids[0]))
        historian_two.append(int(location_ids[1]))

print(historian_one)
print(historian_two)
```

```python
historian_one.sort()
historian_two.sort()

distance_sum = 0

for i in range(len(historian_one)):
    distance_sum += abs(historian_one[i] - historian_two[i])

print(distance_sum)
```

## Part 2

```python
historian_two_freq = {}

for id in historian_two:
    if id in historian_two_freq:
        historian_two_freq[id] += 1
    else:
        historian_two_freq[id] = 1

similarity_sum = 0

for id in historian_one:
    if id in historian_two_freq:
        similarity_sum += id * historian_two_freq[id]

print(similarity_sum)
```
