## Part 1

```python
def load_grid(file_path):
    with open(file_path, 'r') as f:
        return [list(line.strip()) for line in f]

def search_word(grid, word):
    rows, cols = len(grid), len(grid[0])
    word_len = len(word)
    total_matches = 0
    
    # Horizontal, Vertical, and Diagonal Searches
    directions = [
        (0, 1),   # Right
        (0, -1),  # Left
        (1, 0),   # Down
        (-1, 0),  # Up
        (1, 1),   # Down-Right Diagonal
        (1, -1),  # Down-Left Diagonal
        (-1, 1),  # Up-Right Diagonal
        (-1, -1)  # Up-Left Diagonal
    ]
    
    for r in range(rows):
        for c in range(cols):
            if grid[r][c] == word[0]:  # Start matching from the first letter
                for dr, dc in directions:
                    found = True
                    for i in range(word_len):
                        nr, nc = r + dr * i, c + dc * i
                        if not (0 <= nr < rows and 0 <= nc < cols) or grid[nr][nc] != word[i]:
                            found = False
                            break
                    if found:
                        total_matches += 1
                        print(f"Word '{word}' found at ({r}, {c}) in direction ({dr}, {dc})")
    print("Finished searching")
    return total_matches

# Load the grid
grid = load_grid("input.txt")

# Words to search
words_to_search = ["XMAS"]

# Search for each word
for word in words_to_search:
    print(search_word(grid, word))
```
