## Part 1

```python
def load_grid(file_visited):
    with open(file_visited, 'r') as f:
        grid = []
        for line in f:
            grid.append([x for x in line.strip()])
        return grid


def get_area(grid, row, col, region, visited=None):
    if visited is None:
        visited = []

    if row < 0 or row >= len(grid) or col < 0 or col >= len(grid[0]):
        return 0, 1

    if (row, col) in visited:
        return 0, 0

    if grid[row][col] != region:
        return 0, 1

    visited.append((row, col))
    
    valid_area = 1
    perimeter = 0
    up_valid_area,up_perimeter = get_area(grid, row - 1, col, region, visited) 
    down_valid_area,down_perimeter = get_area(grid, row + 1, col, region, visited)
    right_valid_area,right_perimeter = get_area(grid, row, col + 1, region, visited)
    left_valid_area,left_perimeter = get_area(grid, row, col - 1, region, visited)

    valid_area += up_valid_area + down_valid_area + right_valid_area + left_valid_area
    perimeter += up_perimeter + down_perimeter + right_perimeter + left_perimeter

    return valid_area, perimeter

def remove_area(ap_list, item):
    count = 0
    i = 0
    while i < len(ap_list) and count < item:
        if ap_list[i] == item:
            ap_list.pop(i)  # Remove the item and don't increment index
            count += 1
        else:
            i += 1  # Increment index only if no removal
    return ap_list

def find_areas(grid):
    areas = {}
    perimeter = {}
    for row in range(len(grid)):
        for col in range(len(grid[0])):
            if(grid[row][col] not in areas):
                new_area, new_perimeter = get_area(grid, row, col, grid[row][col])
                areas[grid[row][col]] = [new_area]
                perimeter[grid[row][col]] = [new_perimeter]
            else:
                new_area, new_perimeter = get_area(grid, row, col, grid[row][col])
                areas[grid[row][col]].append(new_area)
                perimeter[grid[row][col]].append(new_perimeter)

    unique_areas = {}
    unique_perimeters = {}
            
    for region,areas in areas.items():
        unique_areas[region] = []
        current_area = 0
        for area in areas:
            if current_area != area:
                unique_areas[region].append(area)
                areas = remove_area(areas, area)
                current_area = area
    for region,perimeters in perimeter.items():
        unique_perimeters[region] = []
        current_perimeter = 0
        for perimeter in perimeters:
            if current_perimeter != perimeter:
                unique_perimeters[region].append(perimeter)
                perimeters = remove_area(perimeters, perimeter)
                current_perimeter = perimeter
    
    return unique_areas,unique_perimeters



def calculate_price(area, perimeter):
    price = 0
    for region,areas in area.items():
        perimeters = perimeter[region]
        for i in range(len(areas)):
            price += areas[i] * perimeters[i]
    return price

grid = load_grid("input.txt")

# print(grid)

area , perimeter = find_areas(grid)

print(area)
print(perimeter)
print(calculate_price(area, perimeter))
```
